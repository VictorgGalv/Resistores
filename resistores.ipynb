```python
import sys

print('Utilize apenas letras minusculas.\n')
print('não é nescessário o uso de acentos gramaticais.\n')


loop='s'
while loop=='s':
  faixa= input(str('Esse é um resistor de 4 ou 5 faixas?: '))
  entrada= input(str('você deseja colocar cores ou números como entrada?:'))
  if entrada== ('cores') or ('cor'):
    f1=input(str('digite a cor da primeira faixa: '))
    f2=input(str('digite a cor da segunda faixa: '))
  if faixa== '4' or 'quatro':
      f3= ('  ')
      f4=input(str('digite a cor da terceira faixa (multiplicador): '))
      f5=input(str('digite a cor da quarta faixa (tolerância): '))
    
  elif faixa== '5' 'cinco':
    f3=input(str('digite a cor da terceira faixa: '))
    f4=input(str('digite a cor da quarta faixa (multiplicador): '))
    f5=input(str('digite a cor da quinta faixa (tolerância): '))

  #em caso de os valores colocados serem cores serão executados os seguintes comandos

  #faixa1
  if f1== 'preto':
    n1=(0)
  if f1== 'marrom':
    n1=(1)
  if f1== 'vermelho':
    n1=(2)
  if f1== 'laranja':
    n1=(3)
  if f1== 'amarelo':
    n1=(4)
  if f1== 'verde':
    n1=(5)
  if f1== 'azul':
    n1=(6)
  if f1== 'violeta':
    n1=(7)
  if f1== 'cinza':
    n1=(8)
  if f1== 'branco':
    n1=(9)

  #faixa2
  if f2== 'preto':
    n2=(0)
  if f2== 'marrom':
    n2=(1)
  if f2== 'vermelho':
    n2=(2)
  if f2== 'laranja':
    n2=(3)
  if f2== 'amarelo':
    n2=(4)
  if f2== 'verde':
    n2=(5)
  if f2== 'azul':
    n2=(6)
  if f2== 'violeta':
    n2=(7)
  if f2== 'cinza':
    n2=(8)
  if f2== 'branco':
    n2=(9)

  #faixa3
  if f3== 'preto':
    n3=(0)
  if f3== 'marrom':
    n3=(1)
  if f3== 'vermelho':
    n3=(2)
  if f3== 'laranja':
    n3=(3)
  if f3== 'amarelo':
    n3=(4)
  if f3== 'verde':
    n3=(5)
  if f3== 'azul':
    n3=(6)
  if f3== 'violeta':
    n3=(7)
  if f3== 'cinza':
    n3=(8)
  if f3== 'branco':
    n3=(9)
  if f3== 'preto':
    n3=(0)

  #faixa4 (multiplicador)
  if f4== 'preto':
    n4=('1')
  if f4== 'marrom':
    n4=('x10')
  if f4== 'vermelho':
    n4=('x100')
  if f4== 'laranja':
    n4=('x1k')
  if f4== 'amarelo':
    n4=('x10k')
  if f4== 'verde':
    n4=('x100k')
  if f4== 'azul':
    n4=('x1M')
  if f4== 'violeta':
    n4=('x10M')
  if f4== 'dourado' or 'ouro':
    n4=('x0,1')
  if f4== 'prateado' or 'prata':
    n4=('x0,01')

  #faixa5 (tolerancia)
  if f5== 'marrom':
    n5=('+-1%')
  if f5== 'vermelho':
    n5=('+-2%')
  if f5== 'dourado' or 'ouro':
    n5=('+-5%')
  if f5== 'prateado' or 'prata':
    n5=('+-10%')
  if f5== 'ausencia':
    n5=('+-20%')

  #em caso de os valores colocados de forma numérica serão executados os seguintes codigos

  if entrada== 'numeros' or entrada== 'numero':
    f1=input(str('digite o valor da primeira faixa: '))
    f2=input(str('digite o valor da segunda faixa: '))
  if faixa== ('4'):
    f3= ('  ')
    f4=input(str('digite o valor da terceira faixa (multiplicador): '))
    f5=input(str('digite o valor da quarta faixa (tolerância): '))
    
  elif faixa== ('5'):
    f3=input(str('digite o valor da terceira faixa: '))
    f4=input(str('digite o valor da quarta faixa (multiplicador): '))
    f5=input(str('digite o valor da quinta faixa (tolerância): '))

  #faixa1
  if f1== '0':
    n1=('preto')
  if f1== '1':
    n1=('marrom')
  if f1== '2':
    n1=('vermelho')
  if f1== '3':
    n1=('laranja')
  if f1== '4':
    n1=('amarelo')
  if f1== '5':
    n1=('verde')
  if f1== '6':
    n1=('azul')
  if f1== '7':
    n1=('violeta')
  if f1== '8':
    n1=('cinza')
  if f1== '9':
    n1=('branco')

  #faixa2
  if f2== '0':
    n2=('preto')
  if f2== '1':
    n2=('marrom')
  if f2== '2':
    n2=('vermelho')
  if f2== '3':
    n2=('laranja')
  if f2== '4':
    n2=('amarelo')
  if f2== '5':
    n2=('verde')
  if f2== '6':
    n2=('azul')
  if f2== '7':
    n2=('violeta')
  if f2== '8':
    n2=('cinza')
  if f2== '9':
    n2=('branco')

  #faixa3
  if f3== '0':
    n3=('preto')
  if f3== '1':
    n3=('marrom')
  if f3== '2':
    n3=('vermelho')
  if f3== '3':
    n3=('laranja')
  if f3== '4':
    n3=('amarelo')
  if f3== '5':
    n3=('verde')
  if f3== '6':
    n3=('azul')
  if f3== '7':
    n3=('violeta')
  if f3== '8':
    n3=('cinza')
  if f3== '9':
    n3=('branco')

  #faixa4 (multiplicador)
  if f1== ('1'):
    n1=('preto')
  if f1== '10':
    n1=('marrom')
  if f1== '100':
    n1=('vermelho')
  if f1== '1k':
    n1=('laranja')
  if f1== '10k':
    n1=('amarelo')
  if f1== '100k':
    n1=('verde')
  if f1== '1M':
    n1=('azul')
  if f1== '10M':
    n1=('violeta')
  if f1== '0,1':
    n1=('prateado')
  if f1== '0,001':
    n1=('dourado')

  #faixa5 (tolerancia)
  if f5== '1':
    n5=('marrom')
  if f5== '2':
    n5=('vermelho')
  if f5== '5':
    n5=('dourado')
  if f5== '10':
    n5=('prateado')
  if f5== '20':
    n5=('ausência')

  if entrada== 'numeros':
    if f3== ('  '):
      print('A resistência é {},{},{},{}'.format(n1, n2, n4, n5))
    else:
      print('A resistencia é {},{},{},{},{}'.format(n1, n2, n3, n4, n5))

  if entrada=='cores':
    if f3== ('  '):
      print('A resistência é {}{}{} {} Ω'.format(n1, n2, n4, n5))
  else:
    print('A resistencia é {}{}{}{} {} Ω'.format(n1, n2, n3, n4, n5))

  loop=input('deseja repetir para um novo reistor? s/n\n')
if loop== 'n':
  sys.exit('O programa foi finalizado com sucessoa') 
```
